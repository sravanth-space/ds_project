```python
from tensorflow.keras.models import load_model
import visualkeras
from PIL import ImageFont
```

```python
#  Load Trained Generator
generator = load_model("../saved_models/qpo_cgan_phy_generator.keras")
```

```python
# Customize visualization
try:
    font = ImageFont.truetype("arial.ttf", 12)  # Ensure the font file is available
except IOError:
    font = None  # Fallback if font is not available
```

```python

visualkeras.layered_view(
    generator,
    # to_file="saved_models/qpo_cgan_phy_generator_improved.png",  # Save to file
    min_z=30,  # Increase minimum z size
    min_xy=30,  # Increase minimum x and y size
    max_z=500,  # Increase maximum z size
    max_xy=1000,  # Increase maximum x and y size
    scale_z=0.2,  # Adjust z scaling
    scale_xy=10,  # Adjust x and y scaling
    spacing=50,  # Increase spacing between layers
    draw_volume=True,  # Enable 3D volumetric view
    legend=True,  # Add a legend
    show_dimension=True,  # Show dimensions in the legend
    font=font,  # Use custom font
    font_color="black",  # Set font color
    # color_map={
    #     "InputLayer": {"fill": "cyan", "outline": "black"},
    #     "Dense": {"fill": "orange", "outline": "black"},
    #     "LeakyReLU": {"fill": "pink", "outline": "black"},
    #     "Reshape": {"fill": "purple", "outline": "black"},
    #     "GaussianNoise": {"fill": "green", "outline": "black"},
    #     "GRU": {"fill": "blue", "outline": "black"},
    #     "TimeDistributed": {"fill": "red", "outline": "black"},
    # },
    padding=20,  # Add padding around the visualization
)
```

```python
discriminator = load_model("../saved_models/qpo_cgan_phy_discriminator.keras")
```

```python
from tensorflow.keras.utils import plot_model

# Plot the model architecture
plot_model(discriminator,
           #    to_file='model_plot.png',
           show_shapes=True,
           show_dtype=True,
           show_layer_names=True,
           rankdir='TB',
           expand_nested=True,
           dpi=200,
           show_layer_activations=True,
           show_trainable=False)
```
