```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your val_losses.csv
val_log_path = "saved_models/cgan_hyperparam_runs/run_1_z64_lr0.001_bs64/val_losses.csv"

# Load the validation loss log
val_df = pd.read_csv(val_log_path)

# Check column names to confirm (they may vary depending on your logging)
print(val_df.columns)

# Plot validation generator loss over epochs
plt.plot(val_df['epoch'], val_df['val_gen_loss'], label='Val Generator Loss')
plt.plot(val_df['epoch'], val_df['val_disc_loss'],
         label='Val Discriminator Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Validation Losses over Epochs")
plt.legend()
plt.grid(True)
plt.show()

# Find epoch with lowest generator validation loss
best_epoch = val_df.loc[val_df['val_gen_loss'].idxmin()]
print(f"Best Epoch: {int(best_epoch['epoch'])}")
print(f"Validation Generator Loss: {best_epoch['val_gen_loss']:.4f}")
```
